```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import missingno as msno
```

```python
titanic = sns.load_dataset("titanic")
print(titanic.shape)
titanic.head()
```

```python
msno.matrix(titanic, color=(0.25, 0.25, 0.25))
plt.show()
```

```python
# 색상 변환도 가능하다.
msno.matrix(titanic, color=(0, 0, 1))
plt.show()
```

```python
# bar 형태로도 결측치 확인 가능
msno.bar(titanic)
```

```python
# bar 형태로도 결측치 확인 가능
# bar 형태 또한, 색상변환가능
msno.bar(titanic, color = "darkgreen")
```
