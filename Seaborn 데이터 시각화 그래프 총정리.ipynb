#### 1. Countplot
* 항복별 갯수를 세어주는 countplot
* 알아서 해당 column을 구성하고 있는 value들을 구분하여 보여줌

```python
# 차트 배경 설정
sns.set_style('whitegrid')
# 차트 세로로 그리기
sns.countplot(x="class", hue="who", data = titanic)
plt.show()
```

```python
# 차트 가로로 그리기 : x와 y 변경
sns.countplot(y="class", hue="who", data=titanic)
plt.show()
```

```python
# palette 설정
sns.countplot(x="class", hue = "who", palette = "Accent", data = titanic)
plt.show()
```

#### 2. Displot
* matplotlib의 hist 그래프와 kdeplot을 통합한 그래프
* 분포와 밀도 확인 가능

```python
# displot을 구현하기 위한 sample data 생성
x = np.random.randn(100)
x
```

```python
sns.distplot(x)
plt.show()
```

#### rugplot
* rug는 rugplot이라고도 불리우며, 데이터 위치를 x축 위에 **작은 선분(rug)으로 나타내어
* 데이터들의 위치 및 분포**를 보여준다.

```python
sns.distplot(x, rug=True, hist=False, kde=True)
plt.show()
```

#### kde(kernel density)plot
* kde는 histogram보다 부드러운 형태의 분포 곡선을 보여주는 방법

```python
sns.distplot(x, rug = False, hist = False, kde = True)
plt.show()
```

```python
# 가로로 표현하기
sns.distplot(x, vertical=True)
plt.show()
```

```python
# 컬러 변경하기
sns.distplot(x, color = 'y')
plt.show()
```

#### 3. heatmap
* 색상으로 표현할 수 있는 다양한 정보를 일정한 이미지위에 열분포 형태의 비쥬얼한 그래픽으로 출력 가능하다.

```python
heatmap_data = np.random.rand(10,12)
sns.heatmap(heatmap_data,annot = True)
#annot=True로 해야 수치 표현도 가능
plt.show()
```

```python
#tips data를 pivot table 형태로 변환 후 heatmap으로 표현하기
pivot = tips.pivot_table(index = "day", columns = "size", values = "tip")
pivot
```

```python
sns.heatmap(pivot, annot = True, cmap = "Blues") 
# cmap으로 전체적인 컬러 조정 가능
plt.show()
```

```python
# 상관관계를 표시할 때도 heatmap을 유용하게 사용할 수 있다. 
titanic.corr()
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap = "YlGnBu")
plt.show()
```

#### 4.pairplot
* pairplot은 grid(격자) 형태로 각 집합의 조합에 대해 히스토그램과 분포도를 그린다.
* 또한, 숫자형 column에 대해서만 그려준다.

```python
sns.pairplot(tips)
plt.show()
```

```python
# hue 옵션으로 구분해서 볼수도 있다.
sns.pairplot(tips, hue = 'size')
plt.show()
```

#### 5. violinplot
* 바이올린처럼 생긴 차트 
* column에 대한 데이터의 비교 분포도를 확인할 수 있다
* 곡선진 부분 (뚱뚱한 부분)은 데이터의 분포를 나타낸다
* 양쪽 끝 뾰족한 부분은 데이터의 최소값, 최대값을 나타낸다

```python
sns.violinplot(x=tips['total_bill'])
plt.show()
```

```python
# x, y축을 지정해줌으로써 바이올린을 분할하여 비교 분포를 볼 수 있다
sns.violinplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
# violinplot의 하이라이트로 split 옵션으로 바이올린을 합쳐서 볼 수 있다.
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips, palette="muted",split=True)
plt.show()
```

#### 6. Implot
* Implot은 column 간의 선형관계를 확인하기에 용이한 차트
* 또한, outlier도 같이 짐작해볼 수 있다.

```python
sns.lmplot(x="total_bill", y="tip", data = tips, height=8)
plt.show()
```

```python
# hue 옵션을 사용하여 다중 선형관계 그릴 수 있다
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=8)
plt.show()
```

* 해당 그래프를 통하여 비흡연자가, 흡연자 대비 좀 더 가파른 선형관계를 가지는 것을 볼 수 있다.

```python
# 또한, col 옵션을 추가하여 그래프를 별도로 그려볼 수 있다. 
# col_wrap으로 한 줄에 표기할 column의 갯수를 명시할 수 있다.
sns.lmplot(x='total_bill', y='tip', hue='smoker', col='day', col_wrap=2, height=6, data=tips)
plt.show()
```

#### 7. replot
* 두 column간 상관관계를 보지만 lmplot처럼 선형관계를 따로 그려주지는 않는다.

```python
sns.relplot(x="total_bill", y="tip", hue="day", data=tips)
plt.show()
# 산점도랑 비슷..?
```

```python
# row와 column에 표기할 데이터 column 선택하여 표현 가능
sns.relplot(x="total_bill", y="tip", hue="day", row="sex", col="time", data=tips)
plt.show()
```

#### 8. jointplot
* scatter(산점도)와 histogram(분포)을 동시에 그려주며 숫자형 데이터만 표현 가능

```python
sns.jointplot(x="total_bill", y="tip", height=8, data=tips)
plt.show()
```

```python
# 선형관계를 표현하는 regression 라인 그리기도 가능하다.
# 옵션에 kind='reg'을 추가
sns.jointplot("total_bill", "tip", height=8, data=tips, kind="reg")
plt.show()
```
